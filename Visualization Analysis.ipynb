```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:/Users/Azfar/Desktop/Jupyter Notebook/Intern/Updated-HSB-HKL.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
#print(df.head())

# Check the data types of columns
#print(df.dtypes)

# Get summary statistics
print(df.describe())

# Check for missing values
#print(df.isnull().sum())

# Handle missing values (if needed)
# Example: df.fillna(value, inplace=True)
```

```python
# Example: Histogram of response times

# Example: Bar chart of event types
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="eventCode")
plt.xticks(rotation=90)
plt.xlabel("Event Type")
plt.ylabel("Count")
plt.title("Event Type Distribution")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data=df, x="Distance (km)", bins=20, kde=True, ax=axes[0])
axes[0].set_xlabel("Distance (km)")
axes[0].set_ylabel("Frequency")
axes[0].set_title("Distribution of Distances")

sns.histplot(data=df, x="Duration (mins)", bins=20, kde=True, ax=axes[1])
axes[1].set_xlabel("Duration (mins)")
axes[1].set_ylabel("Frequency")
axes[1].set_title("Distribution of Durations")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Distance (km)', y='Duration (mins)', hue='destinationClinicName', palette='Set1', alpha=0.7, s=100)
plt.xlabel("Distance (km)")
plt.ylabel("Duration (mins)")
plt.title("Distance vs. Duration by Clinic Name")
plt.grid(True)
plt.legend(title='Clinic Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
